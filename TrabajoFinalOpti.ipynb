```python
# Lectura de los dataframes
df_Demanda = pd.read_excel('Distancias.xlsx','Demanda',header=0,index_col=0)
df_costos_R_C = pd.read_excel('Distancias.xlsx','Costos_R_C',header=0,index_col=0)
df_costos_R_M = pd.read_excel('Distancias.xlsx','Costos_R_M',header=0,index_col=0)
df_costos_C_M = pd.read_excel('Distancias.xlsx','Costos_C_M',header=0,index_col=0)
```

```python
model = ConcreteModel(name='Centros')

# Conjuntos
M = list(df_Demanda.index.map(str))          # Municipios
C = list(df_costos_R_C.index.map(str))       # Centros de consolidación

# Parámetros
D = {m: df_Demanda.at[m,'Demanda'] for m in M}                         # Demanda de cada municipio

CostoRC = {c:df_costos_R_C.at[c,"Costos"] for c in C}                  # Costo de Rionegro a centros

CostoRM = {m:df_costos_R_M.at[m,"Costos"] for m in M}                  # Costo de Rionegro a municipios

CostoCM = {(c,m):df_costos_C_M.at[m,c] for c in C for m in M}          # Costo de centros a municipios

T = int(input('¿Cuántos centros de consolidación desea abrir? '))      # Cantidad de centros de consolidación a abrir

# Variables de decisión
model.Y = Var(C, within=Binary)                # Si el centro de consolidación c está abierto
model.W = Var(M, within=Binary)                # Si el municipio m está asignado a Rionegro
model.X = Var(M, C, within=Binary)             # Si el municipio m está asignado a un centro de consolidación
model.Z = Var(C, within=Integers)              # Cantidad de demanda que se enviará a cada centro desde Rionegro
```

```python
# Función objetivo: minimizar costos totales
def obj_rule(model):
    return sum(
        D[m] * model.X[m, c] * CostoCM[c, m]
        for m in M for c in C
    ) + sum(
        D[m] * model.W[m] * CostoRM[m]
        for m in M
    ) + sum(
        model.Z[c] * CostoRC[c]
        for c in C
    )
model.obj = Objective(rule=obj_rule, sense=minimize)

# Restricción 1: Asignación única de municipios
def asignacion_unica_rule(model, m):
    return sum(model.X[m, c] for c in C) + model.W[m] == 1
model.asignacion_unica = Constraint(M, rule=asignacion_unica_rule)

# Restricción 2: Uso de centros de consolidación
def uso_centros_rule(model, m, c):
    return model.X[m, c] <= model.Y[c]
model.uso_centros = Constraint(M, C, rule=uso_centros_rule)

# Restricción 3: Aperturas fijas de centros
def aperturas_centros_rule(model):
    return sum(model.Y[c] for c in C) == T
model.aperturas_centros = Constraint(rule=aperturas_centros_rule)

# Restricción 4: Balance de demanda
def balance_demanda_rule(model, c):
    return model.Z[c] == sum(D[m] * model.X[m, c] for m in M)
model.balance_demanda = Constraint(C, rule=balance_demanda_rule)
```

```python
# Resolver el modelo usando cplex via NEOS
solver_manager.solve(model, opt='cplex')

# Exportar el modelo en formato lp para ver las restricciones
model.write('Centros',format='lp')
```
